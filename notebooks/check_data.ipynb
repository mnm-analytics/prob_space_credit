# value counts

```python
X = train.drop("y", 1)
y = train.y
```

```python
cols_obj = []
cols_flt = []

for col in X.columns:
    if X[col].dtype == object:
        cols_obj.append(col)
    elif X[col].dtype == float:
        cols_flt.append(col)
    else:
        raise Exception
```

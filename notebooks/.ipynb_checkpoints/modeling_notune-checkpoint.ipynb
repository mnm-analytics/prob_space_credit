```python
from load_data import LoadData
from nklearn.lgbm_util.cv import CvEarlyStopping
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
```

# loading

```python
docs = "../docs/"
info = docs + "info/"
finfo = info + "ddl.csv"

data = "../data/"
data_in = data + "in/"
data_out = data + "out/"
fin = data_in + "train_data.csv"
ftest = data_in + "test_data.csv"
fout = data_out + "submit_notune.csv"

loader = LoadData(fin, finfo)
X, y = loader.load()
```

# modeling

```python
import lightgbm as lgb
```

```python
cv = CvEarlyStopping()
```

```python
train, valid = cv.get_lgbd(X, y)
```

```python
X = X.astype(float)
y = y.astype(float)
```

```python
models = cv.fit(X, y, verbose_eval=True)
```

```python
pred = cv.predict(X)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
cv.valid(X, y, roc_auc_score)
```

```python
res = pd.DataFrame({
    "y":y,
    "p":(pred * 100).astype(int)
})
```

```python
res.groupby("p").mean().plot()
```

```python
fpr, tpr, thresholds = cv.valid(X, y, roc_curve)
```

```python
l = (tpr - 1)**2 + fpr**2
```

```python
rc = pd.DataFrame({
    "l":l,
    "thre":thresholds
})
```

```python
thre = float(rc[rc.l == rc.l.min()].thre)
thre
```

```python
plt.scatter(x=fpr, y=tpr)
```

# pred

```python
pd.read_csv(ftest)
```

```python
loader = LoadData(ftest, finfo)
test_X= loader.load()
```

```python
test_X = test_X.astype(float)
```

```python
pred = cv.predict(test_X)
```

```python
pred = pred > thre
```

```python
converter = lambda p: 1 if p else 0
```

```python
ix = test_X.index
```

```python
test_X["Y"] = np.vectorize(converter)(pred)
```

```python
test_X["Y"].to_csv(fout, header=True)
```

# get feature

```python
class GetOnehot:
    
    def fit(self, df_str):
        self.col_names = df_str.columns
        self.vals = [set(df_str[c]) for c in self.col_names]
        
    def get_onehot(self, df):
        cols = {}
        for c, vs in zip(self.col_names, self.vals):
            for v in vs:
                cols[c + "_" + str(v)] = np.where(df[c] == v, 1, 0)
            else:
                cols[c + "_else" ] = np.where(df[c].isin(vs), 0, 1)
        return pd.DataFrame(cols)
```
